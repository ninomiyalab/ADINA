```python
import tensorflow as tf
import tensorflow.keras
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, Dense, Activation, Conv2D, Flatten, Lambda, Resizing, AveragePooling2D
from tensorflow.keras import optimizers
import matplotlib.pyplot as plt
import numpy as np
import csv
from sklearn import datasets
from sklearn.model_selection import train_test_split
from tensorflow.keras import metrics

import sys
sys.path.append('/content/ADINA')
from optimizer.adina import Adina
from optimizer.adinaw import AdinaW

def compare_MNIST(i):
    np.random.seed(i)

    mnist = tf.keras.datasets.mnist
    (x_train, y_train), (x_test, y_test) = mnist.load_data()


    x_train, x_test = x_train[..., np.newaxis]/255.0, x_test[..., np.newaxis]/255.0
    y_train, y_test = tf.keras.utils.to_categorical(y_train), tf.keras.utils.to_categorical(y_test)


    #defined Neural Network Model
    def My_Model(input_shape, Output_dim):
        inputs = Input(shape = (input_shape))
        x = Flatten()(inputs)
        x = Dense(100, activation="relu")(x)
        x = Dense(50, activation="relu")(x)
        output = Dense(Output_dim, activation="softmax")(x)
        model = Model(inputs = [inputs], outputs = [output])
        return model

    model = My_Model(x_train.shape[1:], Output_dim=10)
    model.save("model.h5")

    loss_fn = tf.keras.losses.CategoricalCrossentropy()

    epochs = 50

    mini_batch = x_train.shape[0] #full batch

    #if verbose is True, the results of each iteration will be printed.
    verbose = True

    #if graph is True, the results of all algorithms will be plotted in the graph.
    graph = True

    #-------------------------------------
    p = model(x_train)
    initial_loss = loss_fn(y_train, p)
    initial_accuracy = metrics.Accuracy()(y_train, p)
    p = model(x_test)
    initial_test_loss = loss_fn(y_test, p)
    initial_test_accuracy = metrics.Accuracy()(y_test, p)
    #-------------------------------------

    # Adina (29 April 2025)--------------------------------------------------------------------------------------
    model = load_model("model.h5")

    optimizer = Adina(a=0.1, b = 0.9)

    model.compile(loss=loss_fn, optimizer=optimizer, metrics=['accuracy'])

    ADINA_history = model.fit(x_train, y_train, epochs = epochs, verbose = verbose, batch_size = mini_batch, validation_data = (x_test, y_test))

    # Adina-w (29 April 2025)--------------------------------------------------------------------------------------

    model = load_model("model.h5")

    optimizer = AdinaW(a=0.1, b = 0.9, weight_decay = 0.04)

    model.compile(loss=loss_fn, optimizer=optimizer, metrics=['accuracy'])

    ADINA_W_history = model.fit(x_train, y_train, epochs = epochs, verbose = verbose, batch_size = mini_batch, validation_data = (x_test, y_test))

    # Adam --------------------------------------------------------------------------------------

    model = load_model("model.h5")

    optimizer = keras.optimizers.Adam( )

    model.compile(loss=loss_fn, optimizer=optimizer, metrics=['accuracy'])

    Adam_history = model.fit(x_train, y_train, epochs = epochs, verbose = verbose, batch_size = mini_batch, validation_data = (x_test, y_test))
   # SGD --------------------------------------------------------------------------------------
    model = load_model("model.h5")

    optimizer = keras.optimizers.SGD( )

    model.compile(loss=loss_fn, optimizer=optimizer, metrics=['accuracy'])

    SGD_history = model.fit(x_train, y_train, epochs = epochs, verbose = verbose, batch_size = mini_batch, validation_data = (x_test, y_test))
    # --------------------------------------------------------------------------------------

    # Configure the overall figure with a 2×2 subplot layout.
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle("Training and Testing Metrics", fontsize=16)

    # Plotting data in each subplot

    # ① train_loss
    ax = axes[0, 0]
    ax.plot([initial_loss] + ADINA_history.history['loss'], color="red", label="ADINA")
    ax.plot([initial_loss] + ADINA_W_history.history['loss'], color="pink", label="ADINA_W")
    ax.plot([initial_loss] + Adam_history.history['loss'], color="skyblue", label="Adam")
    ax.plot([initial_loss] + SGD_history.history['loss'], color="green", label="SGD")
    ax.set_title('Train Loss')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Loss')
    ax.legend()

    # ② train_accuracy
    ax = axes[0, 1]
    ax.plot([initial_accuracy] + ADINA_history.history['accuracy'], color="red", label="ADINA")
    ax.plot([initial_accuracy] + ADINA_W_history.history['accuracy'], color="pink", label="ADINAW")
    ax.plot([initial_accuracy] + Adam_history.history['accuracy'], color="skyblue", label="Adam")
    ax.plot([initial_accuracy] + SGD_history.history['accuracy'], color="green", label="SGD")
    ax.set_title('Train Accuracy')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Accuracy')
    ax.legend()

    # ③ test_loss
    ax = axes[1, 0]
    ax.plot([initial_test_loss] + ADINA_history.history['val_loss'], color="red", label="ADINA")
    ax.plot([initial_test_loss] + ADINA_W_history.history['val_loss'], color="pink", label="ADINAW")
    ax.plot([initial_test_loss] + Adam_history.history['val_loss'], color="skyblue", label="Adam")
    ax.plot([initial_test_loss] + SGD_history.history['val_loss'], color="green", label="SGD")
    ax.set_title('Test Loss')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Loss')
    ax.legend()

    # ④ test_accuracy
    ax = axes[1, 1]
    ax.plot([initial_test_accuracy] + ADINA_history.history['val_accuracy'], color="red", label="ADINA")
    ax.plot([initial_test_accuracy] + ADINA_W_history.history['val_accuracy'], color="pink", label="ADINAW")
    ax.plot([initial_test_accuracy] + Adam_history.history['val_accuracy'], color="skyblue", label="Adam")
    ax.plot([initial_test_accuracy] + SGD_history.history['val_accuracy'], color="green", label="SGD")
    ax.set_title('Test Accuracy')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Accuracy')
    ax.legend()

    # Layout adjustment and saving
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    fig.savefig(f'./28MNIST_AllMetrics{i}.png')
    plt.show()


for i in range(1):
    print(i + 1)
    compare_MNIST(i)
```
